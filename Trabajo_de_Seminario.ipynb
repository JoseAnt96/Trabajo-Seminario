```python
import pandas as pd
import io
from google.colab import files
import numpy as np
import itertools
import seaborn as sns
```

```python
def upload_files ():
  uploaded = files.upload()
  for fn in uploaded.keys():
    print('User uploaded file "{name}" with length {length} bytes'.format(
        name=fn, length=len(uploaded[fn])))
    df = pd.read_csv(io.StringIO(uploaded[fn].decode('utf-8')))
    return df
```

```python
estadisticas = upload_files()
estadisticas.head()
```

```python
import plotly.express as px
var = ['Tiros_Campo' , 'Triples', 'Tiros_Libres' , 'Rebotes' , 'Asistencias', 'Robos', 'Tapones', 'Faltas', 'Puntos']

for i in var:
  fig = px.bar(estadisticas, x="NICKNAME", y=i, color="NICKNAME",
  animation_frame="Temporada", animation_group="Temporada", range_y=[0,max(estadisticas[i])])
  print(fig.show())
```
